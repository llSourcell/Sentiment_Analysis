```python
import tweepy
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
#My Twitter API Authentication Variables
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

tweets = api.search('beyonce', count=200)


data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

display(data.head(10))


print(tweets[0].created_at)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()


listy = []

for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
data['polarity'] = se.values

display(data.head(100))
```

```python
type(data['polarity'][4])
data['polarity'][4]
pos = 0
neg = 0
comp = 0
for row in data['polarity']:
    pos += row['pos']
    neg += row['neg']
    comp += row['compound']
    
average_pos = (pos/data['polarity'].size)*100
average_neg = (neg/data['polarity'].size)*100
average_comp = (comp/data['polarity'].size)*100 if (comp/data['polarity'].size)*100 > 0 else 0

print('Positive :',average_pos, 'Negative :', average_neg, 'Compound :', average_comp)
```

```python

plt.bar(height = [average_pos, average_neg, average_comp], x = [ "Good", "Bad", "Compound"])
plt.plot()
```
